```python
def kw_convert(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
      if counter != 3:
        l.append(i["name"])
        counter += 1
        
      else:
         break 
  return l
```

```python
movies['cast']=movies["cast"].apply(kw_convert)
```

```python
movies.head()
```

```python
movies.iloc[0]["crew"]
```

```python
def fetch_director(obj):
  g_list = []
  for o in ast.literal_eval(obj):
    if o["job"] == "Director":
      g_list.append(o["name"])
      break
  return g_list
    

```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ",'') for i in x])
```

```python
movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies = movies[['movie_id','title','tag']]
new_movies.head()
```

```python
new_movies['tag']=new_movies["tag"].apply(lambda x: " ".join(x))
```

```python
new_movies['tag'][0]
```

```python
new_movies['tag']=new_movies['tag'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

## Perform Steming

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem_text(text):
    y = []
    for i in text.split():
        
        y.append(ps.stem(i))
    return " ".join(y)

```

```python
new_movies['tag']=new_movies['tag'].apply(stem_text)
```

## Text Vectorization

We use Bags of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000,stop_words='english')  # keep only top 5000 frequent words
```

```python
vectors = vectorizer.fit_transform(new_movies['tag']).toarray()
vectors.shape
```

```python
vectors[0]
```

```python
np.set_printoptions(threshold=np.inf)

print(vectorizer.get_feature_names_out())
```

## Similarity

cosine similarity measures how alike two vectors are, while cosine distance measures how different they are

We use Cosine Similarity for check how 2 movies are similar

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
print(similarity[0])
```

```python
new_movies.head()
```

```python
def recommend(movie):
	movie_index=new_movies[new_movies['title'] == movie].index[0]
	distance=similarity[movie_index]
	movies_list=sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
	for i in movies_list:
		print(new_movies.iloc[i[0]]['title'])
  
  
```

```python
recommend('Titanic')
```
